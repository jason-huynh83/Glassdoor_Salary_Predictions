# Glassdoor Salary Predictions
- In this notebook, I will be utilizing Machine Learning Algorithms to predict Glassdoor Salaries
- The algorithms used:
    - Multiple Linear Regression
    - Lasso Regression
    - Random Forest Regression

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('glassdoor_jobs_cleaned.csv')
df.head()
```

```python
df.describe().transpose()
```

## Model Build
- choosing relevant columns
- categorical/dummy variables
- outliers
- Train-test-split
- Linear/Lasso Regression (Cross Validation)
- Random Forest Regression
- Cross Validation
- Gridsearch/Hypertune

```python
# Choosing Relevant columns
df2 = df[['Rating','Size','Type of ownership','Industry','Sector','Revenue',
         'Hourly','Employer Provided','Age','in HQ','python_yn','spark_yn',
         'aws_yn','SQL_yn','State','job_simplifier','seniority','desc_len','num_comp',
         'Average Salary']]
```

```python
sns.distplot(df['Average Salary'])
```

- From this distribution plot, we can see there are some outliers near the ends of the plot. We will explore this

```python
# Outlier detection (if any)
max_threshold = df2['Average Salary'].quantile(0.99)
min_threshold = df2['Average Salary'].quantile(0.01)

print('Max threshold at 99% quantile is: {}'.format(max_threshold))
print('Min threshold at 1% quantile is: {}'.format(min_threshold))
```

- Anything above the max_threshold value of 214.735 (99% quantile) we can consider these as outliers
- Anything below the min_threshold value of 39.4 (1% quantile) we can consider these as outliers
- We will explore these values

```python
df2[df2['Average Salary']>max_threshold]
```

- Although these salaries may seem very reasonable for Senior Data Scientists/MLE positions, it may affect the performance of our model, hence we will remove these values from the model to enhance the performance of our model.

```python
df2[df2['Average Salary']<min_threshold]
```

- These values represent more of the outliers in our dataset. A Senior Analyst only making 25k doesn't seem reasonable, as well as a data scientist of an average salary of 15k does not seem reasonable.
- Thus, these values will also be excluded from the dataset to enhance our model performance

```python
df3 = df2[(df2['Average Salary']>min_threshold) & (df2['Average Salary']<max_threshold)]
```

#### Histogram outliers excluded

```python
sns.distplot(df3['Average Salary'])
```

- this dataset will be used for our model building.

```python
# Categorical/dummy variables
df3.head()
```

```python
# Categorical variables will be turned into dummy variables using pd.get_dummies
df_dum = pd.get_dummies(df3)
```

```python
# Train-test-split
from sklearn.model_selection import train_test_split
```

```python
X = df_dum.drop('Average Salary',axis=1)
y = df_dum['Average Salary'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# Multiple Linear Regression
import statsmodels.api as sm

X_sm = X = sm.add_constant(X)
model = sm.OLS(y,X_sm)
model.fit().summary()
```

- This statsmodel represents a lot of relevant and important information.
- The model has a R-squared score of 0.678. This means that the around 68% of the model will fit our multiple linear regression

```python
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.model_selection import cross_val_score
```

```python
lm = LinearRegression()
```

```python
# Cross Validation
cross_val = np.mean(cross_val_score(lm,X_train,y_train,scoring='neg_mean_absolute_error',cv=5))
print('Cross Validation on Linear Model: {}'.format(cross_val))
```

```python
# Fitting X_train, y_train to the linear model
lm.fit(X_train,y_train)
```

```python
# Lasso Regression
lml = Lasso()
```

```python
cross_val_lasso = np.mean(cross_val_score(lml,X_train,y_train,scoring='neg_mean_absolute_error',cv=5))
print('Cross Validation on Lasso Model: {}'.format(cross_val_lasso))
```

```python
lml.fit(X_train,y_train)
```

```python
# Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor()
```

```python
np.mean(cross_val_score(rf,X_train,y_train,scoring='neg_mean_absolute_error',cv=5))
```

```python
# Gridsearch and Hypertuning
from sklearn.model_selection import GridSearchCV
parameters = {'n_estimators':range(10,300,10),'max_features':('auto','sqrt','log2'),'criterion':('mae','mse')}
gs = GridSearchCV(rf,parameters,scoring='neg_mean_absolute_error',cv=5)
gs.fit(X_train,y_train)
```

```python
gs.best_score_
```

```python
gs.best_estimator_
```

```python
# Test predictions
lin_p = lm.predict(X_test)
lasso_p = lml.predict(X_test)
rf_p = gs.best_estimator_.predict(X_test)
```

```python
# Model Evaluation
from sklearn import metrics
print('Multiple Linear Regression MAE: ', metrics.mean_absolute_error(y_test,lin_p))
print('Lasso Regression MAE: ', metrics.mean_absolute_error(y_test,lasso_p))
print('Random Forest Regression MAE: ', metrics.mean_absolute_error(y_test,rf_p))
```

## Conclusion

- From the evaluation we can see that the Lasso Regression Model scores the worst at around 19. This is interesting as I assumed the Lasso would at least perform better than the multiple linear regression as Lasso would help with multicollinearity issues. However, it seems that the Lasso model most likely **overfits** the training set and thus, a higher error in the test set.

- The Random Forest Regressor performs the best with a MAE with around 11. In other words, using the Random Forest Regressor, we are around 11k off in salary predictions.
